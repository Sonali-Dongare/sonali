# BIVARIATE ANALYSIS

```python
plt.figure(figsize=(8,6))
subset_df = df.head(128976).head(30)
sns.scatterplot(x='Region',y='ID',hue='Occupation',data=subset_df,color='Red')
plt.title('Distribution of Region by Occupation')
plt.xlabel('Region')
plt.ylabel('ID')
plt.legend(loc='best')
plt.show()
```

CONCLUSION:

Scatter plot: Scatter plot is used to know the highest sales in which Region based on ID between their Occupation.

# MULTIVARIATE ANALYSIS

```python
# Boxplot for 'Occupation' vs 'Income'
plt.figure(figsize=(10, 6))
plt.xticks(rotation=90)
sns.boxplot(x='Occupation', y='Income', data=subset_df)
plt.title('Occupation vs Income')
plt.show()
```

```python
sns.pairplot(df[['Children', 'Age', 'Cars', 'Income']],diag_kind='kde')
plt.title("Pair plot",y=1.02)
plt.show()
```

Conclusion:

Pairplot:This pair plot visualizes the relationships between the variables 'Children', 'Age', 'Cars' And 'Income'. The scatter plots along the diagonal show the distribution of each variable, while the off-diagonal plots show the relationships between them.The diagonal also includes Kernel Density Estimation (KDE) plots to visualize the probability density of each variable.

Boxplot for Occupation vs Income: The boxplot comparing 'Occupation' and 'Income' visually summarizes how income varies across different occupation categories. It highlights median income, income distribution spread, and potential outliers for each occupation, providing insights into the income patterns associated with different Region.
